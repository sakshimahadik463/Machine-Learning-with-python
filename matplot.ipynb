```python
##you can include the colormap in the drawing by including the plt.colorbar() statement
import matplotlib.pyplot as plt
import numpy as np
x= np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y= np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes=np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
colors=np.array([0,10,20,30,40,45,50,55,60,70,80,90,100])
plt.scatter(x,y,s=sizes,c=colors,cmap='viridis')
plt.colorbar()
plt.show()
```

```python
## you can adjust transperency of the dots with the alpha argument.

## just like colors,make sure the array for sizes has the length as the x-axis and y-axis
x= np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y= np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes=np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
colors=np.array([0,10,20,30,40,45,50,55,60,70,80,90,100])
plt.scatter(x,y,s=sizes,c=colors,alpha=0.3)
plt.show()
```

```python
## if you want to print the bars horizontally instead of vertically ,use the barh() function
x=np.array(['a','b','c','d'])
y=np.array([3,8,1,10])
plt.bar(x,y,color='green')
plt.grid(color='hotpink',linestyle='--',linewidth=1.5)
plt.show()
```

```python
## the barh takes the keyword argument height to set the height of the bars
x=np.array(['a','b','c','d'])
y=np.array([3,8,1,10])
plt.barh(x,y,color='black',height=0.1)
plt.grid(color='green',linestyle='--',linewidth=1.5,alpha=0.3)
plt.show()
```

```python
##add labels to the pie chart with the label parameter
##the label parameter must be an array with one label for ea h wedge

y=np.array([10,25,15,10])
mylabels=['cherries','apples','bananas','dates']
plt.pie(y,labels = mylabels)
plt.show()
```

```python
y=np.array([10,25,15,10])
mylabels=['sakshi','sampada','vasudha','rutuja']
plt.pie(y,labels = mylabels)
plt.show()
```

```python
##as mentioned  the default start angle is at the x-axis,but you can change the start angle 
## by specifying a startangle parameter.

### the startangle parameter is defined with an angle in degree ,default angkle is zero (0):
y=np.array([10,25,15,10])
mylabels=['sakshi','sampada','vasudha','rutuja']
plt.pie(y,labels = mylabels,startangle=90)
plt.show()
```

```python
y=np.array([35,25,25,15])
mylabels=['cherries','apples','bananas','dates']
myexplode=[0,0.95,0.8,0.2]
plt.pie(y, labels=mylabels,explode = myexplode)   ## to make parts  
plt.show()
```

```python
y=np.array([10,25,15,10])
mylabels=['sakshi','sampada','vasudha','rutuja']
mycolors=['black','blue','green','red']
plt.pie(y,labels = mylabels,colors=mycolors,startangle=180)
plt.show()
```

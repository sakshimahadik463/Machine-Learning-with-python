```python
import pandas as pd
import numpy as np
```

```python
labels=['w','x','y','z']
sample_list=[100,200,300,400]
array=np.array([1,2,3,4])
dict={'w':10,'x':20,'y':30,'z':40}
```

```python
pd.Series(data=sample_list)
```

```python
pd.Series(data=array)
```

```python
pd.Series(data=labels)
```

```python
pd.Series(data=sample_list,index=labels)
```

```python
pd.Series(data=sample_list,index=array)
```

```python
pd.Series(sample_list,array)   ##(x,y)----where x=data and y=index
```

```python
pd.Series(dict)
```

```python
pd.Series(array,dict)
```

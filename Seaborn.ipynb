```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/sample_data/train_LZdllcl.csv")
```

```python
df.head()
```

```python
df.head(10)
```

```python
df.sample()
```

```python
df.isnull().sum()
df.describe
```

```python
## there are three kind of data Numerical , Categorical , ??
##for numerical data we use df.relplot() whereas for categorical data we use df.catplot() 
sns.catplot(x="education", y="avg_training_score", data=df)
```

```python
sns.catplot(x="education", y="avg_training_score",jitter=False, data=df)    ##jitter is used to reform the data in single or simplified view
```

```python
sns.catplot(x="education", y="avg_training_score",hue="gender", data=df)
```

```python
sns.catplot(x="education", y="avg_training_score",hue="gender",kind="swarm6", data=df)
```

```python
sns.catplot(x="education", y="avg_training_score",hue = "is_promoted",kind="swarm6", data=df)
```

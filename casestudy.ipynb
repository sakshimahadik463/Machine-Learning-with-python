```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/hbuddana/Cardio-Good-Fitness-Case-Study/master/CardioGoodFitness.csv.txt")
```

```python
sns.boxplot(x="Product",y="Age",data=df) ##Boxplot- one should be numerical
```

```python
pd.crosstab(df["Product"],df["Gender"])
```

```python
sns.pairplot(df,diag_kind="kde")
plt.show
```

```python
corr=df.corr()
corr
```

```python
sns.boxplot(x="Age",data=df)
```

```python
sns.heatmap(corr,annot=True,cmap="turbo") 
plt.show()
```
